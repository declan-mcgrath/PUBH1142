```python
# Background coding

import numpy as np

# Seed the pseudo-random number generator
numpy.random.seed(28)

# Create an array of random values from a normal distribution with a mean of 85 and a standard deviation of 10 and assign it to the variable dBP
dBP = numpy.random.normal(85, 10, 100000)
```

```python
# Calculate the population mean
dBP.mean()
```

> ### 1.

```python
# a random sample of 40 is calculated with the numpy.random.choice function
sample = numpy.random.choice(dBP, 40)
sample
```

> ### 2.

```python
# the sample mean is found with the sample.mean function
sample_mean = sample.mean()
sample_mean
```

> ### 3.

```python
# the sample standard deviation is found with the sample.std function
sample_std = sample.std(ddof=1)
sample_std
```

> ### 4.

```python
# standard error is calculated by dividing the sample standard deviation by the square root of the sample
standard_error = sample_std / numpy.sqrt(40)
standard_error
```

> ### 5.

```python
# scipy.stats must be imported to calculate the confidence coefficient

import scipy.stats as stats

c = stats.t.ppf(0.95, df=39)
c
```

> ### 6.

```python
# the lower bound is caclulated by subtracting the product of the confidence coefficient and standard error from the sample mean
lower_bound = sample_mean - c * standard_error
lower_bound
```

> ### 7.

```python
# the upper bound is calculated by multiplying the confidence coefficient by standard error, then adding the sample mean
upper_bound = sample_mean + (c * standard_error)
upper_bound
```

> ### 8.

```python
# the np.array function is used to create the bootstrapped array of 1000 values
bootstrapped_means = np.array([np.random.choice(sample, size=40, replace=True).mean() for _ in range(1000)])
bootstrapped_means
```

> ### 9.

```python
# the np.percentile function is used to find the 2.5 percentile in the bootstrapped values
bootstrapped_2_5_percent = np.percentile(bootstrapped_means, 2.5)
bootstrapped_2_5_percent
```

> ### 10.

```python
# the np.percentile function is used again to find the 97.5 percentile in the bootstrapped values
bootstrapped_97_5_percent = np.percentile(bootstrapped_means, 97.5)
bootstrapped_97_5_percent
```
